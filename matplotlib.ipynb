```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\ADMIN\Downloads\dataset\Crop_recommendation.csv")
```

```python
humidity=df['humidity']
ph=df['ph']
plt.plot(humidity, ph, marker='o')
plt.title("humidity vs ph")
plt.xlabel("humidity")
plt.ylabel("ph")
plt.show()
```

```python
dept_avg = df.groupby('temperature')['ph'].mean()

plt.bar(dept_avg.index, dept_avg.values)
plt.title("temperaturevs ph")
plt.xlabel("temperature")
plt.ylabel("ph")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(df['ph'], df['rainfall'])
plt.title("ph vs rainfall")
plt.xlabel("ph")
plt.ylabel("rainfall")
plt.show()
```

```python
plt.hist(df['rainfall'], bins=5)
plt.title("rainfall in land")
plt.xlabel("rainfall")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(df['N'])
plt.title("Training Score Box Plot")
plt.ylabel("N")
plt.show()
```

```python
promo = df['ph'].value_counts()
plt.pie(promo, labels=promo.index, autopct='%1.1f%%')
plt.title("Ph level")
plt.show()
```

```python
plt.plot(df['temperature'], df['ph'], label='Ph level')
plt.plot(df['rainfall'], df['N'], label='Previous level')
plt.legend()
plt.title("Performance Metrics by Age")
plt.xlabel("ph")
plt.ylabel("N")
plt.show()
```
